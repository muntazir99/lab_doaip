```python
# Importing required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import GradientBoostingClassifier
from lazypredict.Supervised import LazyClassifier
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
df=pd.read_csv("/content/drive/MyDrive/country_wise_latest.csv")
```

```python
df.head()
```

```python
# Null values
df.isna().sum()
```

```python
df.info()
```

```python
# Changing name of a feature
df.rename(columns={'Country/Region':'Country'},inplace=True)
```

```python
# All countries
df.Country.unique()
```

```python
df['Deaths / 100 Recovered'].value_counts()
```

```python
# Removing Countries and inf values
df.drop(['Country','Deaths / 100 Recovered'],axis=1,inplace=True)
```

```python
df['WHO Region'].unique()
```

```python
# Label Encoding Target Column
le=LabelEncoder()
df['WHO Region']=le.fit_transform(df['WHO Region'])
```

```python
df.head()
```

```python
df.info()
```

```python
# Choosing dependent and independent variables
X=df.drop('WHO Region',axis=1)
y=df['WHO Region']
```

```python
# Splitting data to train test data
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=43)
```

```python
X_train
```

```python
y_train
```

```python
# Selecting model
mod = LazyClassifier(verbose=0,ignore_warnings=True, custom_metric=None)
models,predictions = mod.fit(X_train, X_test, y_train, y_test)
print(models)
```

```python
# Training model
clf=GradientBoostingClassifier()
clf.fit(X_train,y_train)
```

```python
# Prediction
pred=clf.predict(X_test)
```

```python
# Accuracy score
accuracy_score(pred,y_test)
```

```python
# Confusion matrix
confusion_matrix(pred,y_test)
```
